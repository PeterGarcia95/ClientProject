```python
import numpy as np
import pandas as pd
import geopandas as gpd
import scipy
import shapely
import matplotlib.pyplot as plt
```

```python
geo_df = gpd.read_file("../datasets/ZIP_Codes/ZIP_Codes.shp").to_crs(epsg=4326)
```

```python
geo_df.head()
```

```python
geo_df.dtypes
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
geo_df.plot(ax = ax);
ax.set_title('Boston Zipcodes', fontsize = 20);
```

```python
clean_tweets_df = pd.read_csv('../datasets/tweets_and_preds.csv')
clean_tweets_df
```

```python
clean_tweets_df["target"].value_counts()
```

```python
preds = pd.read_csv('../datasets/preds_index.csv')
preds
```

```python
index_to_check = set(preds["index"])
```

```python
# clean_tweets_df[clean_tweets_df["Unnamed: 0"] == 1489]
```

```python
accuracy_map = pd.concat([clean_tweets_df, preds])
```

```python
clean_tweets_df.drop( columns= "targets", axis = 1 , inplace = True)
```

```python
clean_tweets_df
```

```python
accuracy_df = [["target", "preds", "origin"]]
accuracy_df
```

```python
accuracy_map["correct"] = ['True' if accuracy_map.loc[i, 'target'] == accuracy_map.loc[i, 'preds'] else 'False' for i in accuracy_map.index]
```

```python
accuracy_map
```

```python
accuracy_df["col"] = ['True Prediction' if accuracy_df.loc[i, 'target'] == accuracy_df.loc[i, 'preds'] else 'False Prediction' for i in accuracy_df.index]
```

```python
accuracy_df
```

```python
accuracy_gdf = gpd.GeoDataFrame(accuracy_map, 
                            geometry=gpd.points_from_xy(accuracy_map.long, 
                                                        accuracy_map.lat))
```

```python
accuracy_gdf
```

```python
clean_gdf = gpd.GeoDataFrame(clean_tweets_df, 
                            geometry=gpd.points_from_xy(clean_tweets_df.long, 
                                                        clean_tweets_df.lat))
```

```python
clean_gdf.head()
```

```python
clean_gdf.plot(figsize = (20,20));
plt.title('Tweets in Boston', fontsize = 20);
plt.xlabel('Longitude', fontsize = 10);
plt.ylabel('Latitude', fontsize = 10);
```

```python
clean_pred = accuracy_gdf[accuracy_gdf['geometry'].within(geo_df['geometry'].unary_union)]
# only keep those points within polygons
boston_points = clean_gdf[clean_gdf['geometry'].within(geo_df['geometry'].unary_union)]
```

```python
color_dict = {1: 'Outage',0:'Not Outage'}
color_map = {1: 'Outage',0:'Not Outage'}
clean_pred["color"] = clean_pred["target"].map(color_map)
boston_points['color'] = boston_points['target'].map(color_dict)
```

```python
fig, ax = plt.subplots(figsize = (20,20))
geo_df.plot(ax = ax, facecolor = 'aliceblue', edgecolor = 'indigo');
clean_gdf.plot(ax = ax);
ax.set_title('Tweets in Boston', fontsize = 20)
ax.set_facecolor('grey')
ax.set_xlabel('Longitude', fontsize = 16);
ax.set_ylabel('Latitude', fontsize = 16);
```
