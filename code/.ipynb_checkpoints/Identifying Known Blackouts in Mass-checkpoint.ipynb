```python
import pandas as pd
```

Power Outages 2017

```python
outages_2017 = pd.read_csv('../datasets/raw_data/2017_Annual_Summary.csv')
```

```python
outages_2017
```

```python
outages_2017.shape
```

```python
outages_2017.drop(columns= ["Unnamed: 11", "Unnamed: 12", "Unnamed: 13", "Unnamed: 14"], inplace = True)
```

```python
outages_2017.shape
```

```python
outages_2017.dropna(how = 'all', inplace = True)
```

```python
outages_2017 = outages_2017[(outages_2017["Event Type"] == "Severe Weather") & (outages_2017["Area Affected"].str.contains("aine")) ]
```

```python
outages_2017
```

We only have tweets beginning from September 2017. We will only consider blackouts caused by severe weather after September 2017, in Maine. 

```python
outages_2017 = outages_2017[outages_2017["Month"] == "October"]
outages_2017
```

Power Outages 2018 

```python
outages_2018 = pd.read_csv('../datasets/raw_data/2018_Annual_Summary.csv')
outages_2018
```

```python
outages_2018.shape
```

```python
outages_2018.dropna(how = 'all', inplace = True)
```

```python
outages_2018 = outages_2018[(outages_2018["Event Type"] == "Severe Weather") & (outages_2018["Area Affected"].str.contains("aine")) ]
outages_2018
```

Power Outages 2019 

```python
outages_2019 = pd.read_csv('../datasets/raw_data/2019_Annual_Summary.csv', header = 1)
outages_2019
```

```python
outages_2019.shape
```

```python
outages_2019.dropna(how = 'all', inplace = True)
```

```python
outages_2019 = outages_2019[(outages_2019["Event Type"] == "Severe Weather") & (outages_2019["Area Affected"].str.contains("aine")) ]
outages_2019
```

```python
outages_2019.shape
```

Joining out Data 

```python
outages_2017_18 = pd.concat([outages_2017, outages_2018])
```

```python
outages = pd.concat([outages_2017_18, outages_2019])
```

Let's make sure that we have the time in DateTime format

```python
outages.loc[:,"Date Event Began"] = pd.to_datetime(outages["Date Event Began"])
```

```python
outages.loc[:,"Date of Restoration"] = pd.to_datetime(outages["Date of Restoration"])
```

```python
outages
```

```python
outages.dtypes
```

```python
outages.to_csv("../datasets/Outages_since_2017.csv")
```
